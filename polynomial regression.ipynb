```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('/content/Position_Salaries.csv')
df
```

```
X = df.iloc[:,1:-1].values
y= df.iloc[:,-1].values
```

linear regression

```
from sklearn.linear_model import LinearRegression
```

```
lr = LinearRegression()
```

```
lr.fit(X,y)
```

polinomial regression

```
from sklearn.preprocessing import PolynomialFeatures
plr = PolynomialFeatures(degree = 4)
X_polynomial = plr.fit_transform(X)
```

```
lr_poly = LinearRegression()
lr_poly.fit(X_polynomial,y)
```

```
plt.scatter(X,y,color='blue')
plt.plot(X, lr.predict(X) , color = 'red')
plt.title("predictions try - lr")
plt.xlabel('true value')
plt.ylabel('pred value')
plt.show()
```

```
plt.scatter(X,y,color='blue')
plt.plot(X, lr_poly.predict(X_polynomial ), color = 'red')
plt.title("predictions try - poly")
plt.xlabel('true value')
plt.ylabel('pred value')
plt.show()
```

```
y_pred_linear = lr.predict([[6.5]])
y_pred_linear
```

```
y_pred_poly = lr_poly.predict(plr.fit_transform([[6.5]]))
y_pred_poly
```
